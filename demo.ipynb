Demo of PySpark in VSCode

```python
from pyspark.sql import SparkSession
```

Start Spark

```python
spark = SparkSession.builder.master('local[*]').getOrCreate()

print(f'Spark version: {spark.version}')
print(f'Spark UI available at: http://localhost:4040')
```

Create some sample data

```python
df = spark.createDataFrame([
    ('Alice', 34, 'Data Engineer'),
    ('Bob', 45, 'Data Scientist'),
    ('Charlie', 29, 'ML Engineer'),
    ('Diana', 38, 'Software Engineer')
], ['name', 'age', 'role'])
df.show()
```

Select some rows and display the result using Pandas

```python
df.where("role LIKE '%Engineer%'").toPandas()
```
